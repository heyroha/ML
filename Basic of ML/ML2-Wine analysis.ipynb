## 레드/화이트 와인 분류기

```python
x = wine.drop(['color'], axis = 1)
y = wine['color']
```

```python
from sklearn.model_selection import train_test_split
import numpy as np

x_tr,x_t,y_tr,y_t = train_test_split(x, y, test_size = 0.2, random_state= 0xC0FFEE)

np.unique(y_tr, return_counts = True)
```

```python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Histogram(x = x_tr['quality'], name = 'Train'))
fig.add_trace(go.Histogram(x = x_t['quality'], name = 'Test'))

fig.update_layout(barmode ='overlay')
fig.update_traces(opacity = 0.75)
fig.show()
```

```python
from sklearn.tree import DecisionTreeClassifier

wine_tree = DecisionTreeClassifier(max_depth = 2, random_state = 12)
wine_tree.fit(x_tr,y_tr)
```

```python
from sklearn.metrics import accuracy_score

y_pred = wine_tree.predict(x_tr)
y_pred_test = wine_tree.predict(x_t)

print(f'Train acc: {accuracy_score(y_tr, y_pred)}')
print(f'Test acc: {accuracy_score(y_t, y_pred_test)}')
```

```python
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler

estimators = [
    ('scaler',StandardScaler()),
    ('clf', DecisionTreeClassifier())
]
pipe = Pipeline(estimators)
```

```python
pipe
```

```python
pipe.steps[0]
```

```python
pipe['scaler']
```

```python
pipe.set_params(clf__max_depth = 2)
pipe.set_params(clf__random_state = 0xC0FFEE)
```

```python
from sklearn.model_selection import train_test_split

x_tr,x_t,y_tr,y_t = train_test_split(x,y,
                                     test_size = 0.2,
                                     random_state = 0xC0FFEE,
                                     stratify = y)
```

```python
x_tr
```

```python
pipe.fit(x_tr, y_tr)
```

```python
from sklearn.metrics import accuracy_score

y_pred_tr = pipe.predict(x_tr)
y_pred_test = pipe.predict(x_t)

print(f'acc_tr: {accuracy_score(y_tr, y_pred_tr)}')
print(f'acc_t: {accuracy_score(y_t, y_pred_test)}')
```
